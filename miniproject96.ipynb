##  Identifying and Removing Duplicates 

///Objective: Clean a dataset by identifying and removing both exact and partial duplicates. //

```python
import pandas as pd
```


///Load a Dataset://

```python
data = {
    'name': ['John Doe', 'John Doe', 'Alice', 'ALICE ', 'Bob', 'Bob', 'Charlie', 'charlie'],
    'age': [25, 25, 30, 30, 22, 22, 28, 28],
    'city': ['NY', 'NY', 'LA', 'LA', 'TX', 'TX', 'NY', 'NY']
}
df=pd.DataFrame(data)
print("Original data is :\n",df)
```

//// Identify Exact Duplicates:

```python
dupli=df.duplicated()  ##(case-sentitive)
print("Identify the duplicates:\n",dupli)
num_dupli=df.duplicated().sum()
print("Number of duplicates:\n",num_dupli)
```

////Remove Exact Duplicates: //

```python
rem_dupli=df.drop_duplicates(inplace=True)
print("After removing duplicates from the dataset:\n",df)
```

////Handle Fuzzy Duplicates:///

```python
df['name']=df['name'].str.lower().str.strip()
print("After cleaning fuzzy data:\n",df)
```

```python
count=df.groupby('name').size().sort_values(ascending=False)
print("count the name using groupby function:\n",count)
```

///Final check://

```python
final_dupli=df.drop_duplicates(inplace=True)
print("Data After removing fuzzy Duplicates:\n",df)
output=df.duplicated().sum()
print("\n Final Output is:\n",output)
```
