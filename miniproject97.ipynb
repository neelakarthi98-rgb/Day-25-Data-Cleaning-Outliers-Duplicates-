## Sales Transaction Outlier Detection

////Objective: Find and analyze outliers in a sales transaction dataset (e.g., transactions that are unusually high or low).///

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

///Load data://

```python
np.random.seed(42)
sales_data=np.random.normal(loc=500,scale=50,size=200).tolist()
sales_data.extend([1500,2000,2500,50])  ##set outliers
df=pd.DataFrame({"total_sales":sales_data})
print("Original data is:\n",df.head())
```

////Visualize to show the outliers://

```python
plt.figure(figsize=(8, 5))
plt.hist(df['total_sales'], bins=30)
plt.title("Histogram of Total Amount Before handling outliers")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```


////calculate Z-score//

```python
mean_amount=df['total_sales'].mean()
std_amount=df['total_sales'].std()
df['z-score']=(df['total_sales']-mean_amount)/std_amount
print("\n After calculating Z-score:\n",df.head())
```

////Identify Outlier using Z-score

```python
outlier=df[np.abs(df['z-score'])>3]
print("Detected outliers:\n",outlier)
```

///Count the outliers://

```python
count=outlier.shape[0]
print("count the number of outliers:\n",count)
```
